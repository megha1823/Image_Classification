```python
from bing_image_downloader import downloader
downloader.download("icecream cone",limit = 30,output_dir = "images",adult_filter_off = True)
```

```python
downloader.download("cats",limit = 30,output_dir = "images",adult_filter_off = True)
```

```python
downloader.download("cars",limit = 30,output_dir = "images",adult_filter_off = True)
```

```python
#Preprossessing of the data
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

output = []
images = []
flattened_data = []

DATADIRECTORY = '/content/images'
CATEGORIES = ['icecream cone','cats','cars']

for category in CATEGORIES:
  #label encoding of the values
  class_num = CATEGORIES.index(category)
  #creating path to use all images
  path = os.path.join(DATADIRECTORY,category)
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    #resizing the data
    img_resized =resize(img_array,(150,150,3))
    #flatteninf the data
    flattened_data.append(img_resized.flatten())
    images.append(img_resized)
    output.append(class_num)
flattened_data = np.array(flattened_data)
output = np.array(output)
images = np.array(images)
```

```python
#Data Splitting into training data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flattened_data,output,test_size = 0.3,random_state = 109)
```

```python
from sklearn import svm 
clf=svm.SVC(kernel='linear',gamma='auto',C=2)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```
